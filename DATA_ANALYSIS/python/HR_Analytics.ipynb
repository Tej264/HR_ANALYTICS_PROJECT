```python
df['BusinessTravel'] = df['BusinessTravel'].replace('Travel_Rarely', 'TravelRarely')
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
#correlation of attrition to other columns
correlation = df.corr(numeric_only=True)['Attrition']
```

```python
print(correlation)
```

```python
import matplotlib.pyplot as plt
```

```python
correlation = correlation.drop('Attrition')
```

```python
plt.figure(figsize=(8,5))
correlation.sort_values().plot(kind='barh', color='skyblue')

plt.title("Correlation of Features with Attrition")
plt.xlabel("Correlation Value")
plt.ylabel("Features")
plt.show()
```

```python
import seaborn as sns
```

```python
top_cols = ['TotalWorkingYears', 'JobLevel', 'YearsInCurrentRole', 'MonthlyIncome', 'Age']
plt.figure(figsize=(18, 12))

# Loop through each column and create a boxplot
for i, col in enumerate(top_cols):
    plt.subplot(3, 2, i+1)  # 3 rows, 2 columns grid
    sns.boxplot(x='Attrition', y=col, data=df)
    plt.title(f'Relation between {col} and Attrition')

plt.tight_layout()
plt.show()
```
